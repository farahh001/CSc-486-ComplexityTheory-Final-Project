```python
import matplotlib.pyplot as plt
import numpy as np

# Average stock prize in last 5 years
original_image_size = [121.85, 67.19, 96.46, 86.567, 24.874, 49.518, 146.61]
encrypted_image_size = [163, 168, 167, 147, 29, 55, 157]
encrypted_image_size_chaos_rsa = [182, 84, 187, 181, 45, 54, 170]

sample_data = ['original_image.png', 'flower.png', 'dog.png', 'scene.png', 'amazon.png', 'docs.png', 'our_campas.png']

# We cannot add width to year so we create another list
indices = np.arange(len(sample_data))

width = 0.20

# Plotting
plt.bar(indices, original_image_size, width=width)

# Offsetting by width to shift the bars to the right
plt.bar(indices + width, encrypted_image_size, width=width)
plt.bar(indices + width + width, encrypted_image_size_chaos_rsa, width=width)


# Displaying year on top of indices
plt.xticks(ticks=indices, labels=sample_data)

plt.xticks(rotation=45, ha='right')

plt.xlabel("Sample data")
plt.ylabel("Image Size (KB)")
plt.title("Original Image Size vs. Encrypted Image size RSA\nvs. Encrypted Image Size Chaos-based RSA")

plt.savefig("plot.png")
plt.show()


```

```python
from matplotlib import rcParams
size = [24.874, 49.518, 67.19, 86.567,96.46,121.85,146.61]
rsa_encrypted_time = [2.74,4.052, 3.09,4.214,3.872,3.192,4.028]
chaos_ras_encrypted_time = [2.375,2.363,2.377,2.347,2.341,1.625,2.377]
plt.title("RSA Encrypted Time vs. Chaos-based RSA Encrypted Time")
plt.xlabel("Image Size (KB)")
plt.ylabel("Time (Second)")
plt.plot(size, rsa_encrypted_time, label = 'RSA Encrypted Time')
plt.plot(size, chaos_ras_encrypted_time, label = 'Chaos-based RSA Encrypted Time')
plt.legend()
```
